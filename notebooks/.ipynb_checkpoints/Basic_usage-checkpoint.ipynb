# Demo Usage on 2D toy data

```python
import numpy as np
from finch import FINCH
```

# Load Gestalt 2D toy clustering data

We load the 2D Aggregation clustering dataset which has 399 samples belonging to 6 classes. The data is included in the github repo in the data folder.

```python
data_path = "../data/toy_data/gestalt_clusters.txt"
data = np.genfromtxt(data_path)
gt = data[:, 2]
data = data[:, :2]
```

```python
data.shape
```

```python
# ground truth clusters
len(np.unique(gt))
```

# Cluster with FINCH

Since here the the toy data has 2d coordinates specifying the position of points, we cluster it using euclidean distance. Note that the default distance is cosine which is recommended for real-world data.

```python
c, num_clust, req_c = FINCH(data, distance='euclidean')
```

### Required number of clusters

In many application we need to cluster a dataset to obtain a specifioed number of clusters. This can be done easily with FINCH by setting the input param req_clust. For example lets cluster the dataset to obtain the required 7 cluster partition.

Here the variable req_c will contain the labels of the 7 clusters.

```python
c, num_clust, req_c = FINCH(data, req_clust=len(np.unique(gt)), distance='euclidean', verbose=False)
```

```python
print(req_c.shape)
print(np.unique(req_c))
```

## Evaluate quality of clustering

```python
from sklearn.metrics import normalized_mutual_info_score as nmi
score = nmi(gt, req_c)
print('NMI Score: {:.2f}'.format(score * 100))
```

## Visualize the returned clustering labels

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 2, figsize=(10,5))

ax[0].set_title('GT data')
ax[0].scatter(*data.T, s=50, c=gt, cmap='Spectral')
ax[1].set_title('clustering result')
ax[1].scatter(*data.T, s=50, c=req_c, cmap='Spectral')

ax[0].set_xticks([])
ax[0].set_yticks([])
ax[1].set_xticks([])
ax[1].set_yticks([])
plt.show()
```
